```python
# 1.7. Tworzenie wykresów

# Część A (2 punkty)
# Wygeneruj wykres funkcji y=f(x)=(a−4)x^2+(b−5)x+(c−6) , gdzie a,b,c to trzy ostatnie cyfry Twojego numeru indeksu.
# Pamiętaj o opisaniu osi wykresu (x oraz y).

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define constants based on student's ID (361384)
a = 3
b = 8
c = 4

# Initialize
fig = plt.figure()
# Add subplot
ax = fig.add_subplot(111)
plt.style.use('ggplot')

ax.set_xlabel('x')  # X axis label
ax.set_ylabel('y')  # Y axis label
ax.set_title('Wykres funkcji')  # Title

p = -(b-5) / (2 * (a-4)) # Vortex of the parabola -b/2a

x = np.arange(p-20, p+20, 0.1)  # Define x

# calculate y values
y = (a-4) * (x**2) + (b-5) * x + (c-6)

ax.plot(x, y, color='blue', lw=2)
plt.show()  # Show plot

```

```python
# Część B (1 punkt)
# Na wykresie z zadania 1.7A przedstaw dodatkowo wykres funkcji y=g(x)=exex+1.
# Linie przestawiające funkcje y=f(x) i y=g(x) powinny znaleźć się na tym samym wykresie, ale powinny różnić się np. kolorami.

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define constants
a = 3
b = 8
c = 4

# Initialize
fig = plt.figure()
# Add subplot
ax = fig.add_subplot(111)

ax.set_xlabel('x')  # X axis label
ax.set_ylabel('y')  # Y axis label
ax.set_title('Wykres funkcji')  # Title

x = np.arange(0, 2, 0.01)  # Define x

# calculate y values
y = (a-4)*(x**2) + (b-5)*x + (c-6)
y2 = np.exp(x)/(np.exp(x)+1)

ax.plot(x, y, color='blue', lw=2)
ax.plot(x, y2, color='red', lw=2)

plt.show()  # Show plot
```
